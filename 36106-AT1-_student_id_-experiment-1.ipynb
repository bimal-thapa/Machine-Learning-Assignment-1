## E. Train Machine Learning Model

```
train_model_executive_summary = 'Several multivariate linear regression models were trained using different values of the fit_intercept parameter. This was done to see the performance of this model in predicting salaries with different hyperparameters. The results will later be compared with the baseline to find out if there is any improvement.'
```

### E.1 Import Algorithm

> Rationale: We implement linear regression to train models and further evaluate their predictive capability with respect to salaries. This algorithm is simple, interpretable, and very effective in establishing the linear relationship of features to the target variable, which makes it ideal in this case and can give a solid foundation for model experimentation in finding a better baseline.

```
# <fill_this>
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

<hr>

### E.2 Set Hyperparameters

> Rationale: This would be altered by tuning such hyperparameters as fit_intercept. The possibilities given for this hyperparameter would, hence, provide the capability to test different results of the model settings. The models will be trained with and without intercept inclusions, and analysis will be performed to understand which setting better fits the underlying data to minimize prediction error with an aim of fine tuning the model for better results.

```
# <fill_this>
# Initialize models with different values for the hyperparameter fit_intercept
model_intercept = LinearRegression(fit_intercept=True)
model_no_intercept = LinearRegression(fit_intercept=False)
```

<hr>

### E.3 Fit Model

```
# Fit the models with X_train and y_train
model_intercept.fit(X_train[features_list], y_train)
model_no_intercept.fit(X_train[features_list], y_train)

# Predict on validation data for evaluation later
y_pred_intercept = model_intercept.predict(X_val[features_list])
y_pred_no_intercept = model_no_intercept.predict(X_val[features_list])

```

<hr>

### E.4 Model Technical Performance

```
# <fill_this>
# Predict on test data for both models
y_test_pred_intercept = model_intercept.predict(X_test[features_list])
y_test_pred_no_intercept = model_no_intercept.predict(X_test[features_list])

# Evaluate model performance for the model with intercept
print("Model with Intercept Performance:")

# RMSE
rmse_val_intercept = mean_squared_error(y_val, y_pred_intercept, squared=False)
rmse_test_intercept = mean_squared_error(y_test, y_test_pred_intercept, squared=False)
print(f"Validation RMSE: {rmse_val_intercept}")
print(f"Test RMSE: {rmse_test_intercept}")

# Evaluate model performance for the model without intercept
# RMSE
print("Model with No Intercept Performance:")
rmse_val_no_intercept = mean_squared_error(y_val, y_pred_no_intercept, squared=False)
rmse_test_no_intercept = mean_squared_error(y_test, y_test_pred_no_intercept, squared=False)
print(f"Validation RMSE: {rmse_val_no_intercept}")
print(f"Test RMSE: {rmse_test_no_intercept}")

# # Calculate RMSE for both models
# rmse_intercept = mean_squared_error(y_val, y_pred_intercept, squared=False)  # RMSE
# rmse_no_intercept = mean_squared_error(y_val, y_pred_no_intercept, squared=False)  # RMSE

# # Print RMSE results for both models
# print(f"RMSE with intercept: {rmse_intercept}")
# print(f"RMSE without intercept: {rmse_no_intercept}")
```

> Results: The model with an intercept achieved a validation RMSE of 5029.29 and a test RMSE of 15767.78, indicating a significantly better fit compared to the model without an intercept, which had a validation RMSE of 34349.01 and a test RMSE of 38916.51. This suggests that including the intercept improves the model's accuracy and generalizability across both validation and test datasets.

### E.5 Business Impact from Current Model Performance

```
# Ensure that mean_salary is a scalar value
mean_salary = y_train.mean()  # This should work fine unless y_train contains NaN values

# Impact of RMSE
print(f"Mean salary in the training set: {mean_salary}")

# Convert rmse_test_intercept and rmse_test_no_intercept to float if they are not already scalar values
rmse_test_intercept = float(rmse_test_intercept)  # Convert to float if needed
rmse_test_no_intercept = float(rmse_test_no_intercept)  # Convert to float if needed

# Check if mean_salary is a scalar
mean_salary = float(mean_salary)  # Convert to float if it's a Series or non-scalar

# Compare the error margins (RMSE) to the mean salary
print(f"RMSE with intercept accounts for {rmse_test_intercept/mean_salary:.2%} of the average salary")
print(f"RMSE without intercept accounts for {rmse_test_no_intercept/mean_salary:.2%} of the average salary")

# Business Interpretation:
# If RMSE is high compared to the mean salary, then the model's salary predictions might not be very useful.

# Recommendations:
if rmse_test_intercept > mean_salary * 0.1:  # If error is more than 10% of average salary
    print("Recommendation: The error is too high. Consider feature engineering or different algorithms.")
else:
    print("Recommendation: The model is performing well; further refinements may improve business impact.")
```

> Results: The mean salary in the training set is $85,715.32. The RMSE for the model with an intercept accounts for 18.40% of the average salary, while the RMSE for the model without an intercept represents 45.40% of the average salary. Both models show substantial error relative to the average salary, indicating that the current model performance is inadequate and suggesting a need for further feature engineering or alternative algorithms to improve accuracy.
